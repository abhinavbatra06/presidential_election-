```python
from census import Census
import pandas as pd

# Initialize Census object with API key
api_key = ""  # Replace with your Census API key
c = Census(api_key)

# Function to fetch ACS data for specific year and variables
def fetch_acs5_data(year, variables):
    # Retrieve state-level data
    data = c.acs5.state(
        variables,
        state_fips="*",  # Fetch data for all states
        year=year
    )
    # Convert to DataFrame
    return pd.DataFrame(data)

# Define variables
variables = {
    "B19013_001E": "Median Household Income",  # Median household income
    "B23001_012E": "Male Unemployed",          # Male unemployed (16-64 years)
    "B23001_037E": "Female Unemployed",         # Unemployed
    "B23001_007E": "Male Labor Force",         # Male in labor force (16-64 years)
    "B23001_032E": "Female Labor Force"        # Female in labor force (16-64 years)

}

# Fetch data for each year from 2000 to 2023
years = range(2009, 2022)
all_data = []

for year in years:
    try:
        data = fetch_acs5_data(year, list(variables.keys()))
        data = data.rename(columns=variables)
        data["Year"] = year
        all_data.append(data)
    except Exception as e:
        print(f"Error fetching data for {year}: {e}")

# Combine all years into a single DataFrame
acs_data = pd.concat(all_data, ignore_index=True)

# Save to CSV
acs_data.to_csv("acs5_state_data_2000_2023.csv", index=False)
```
